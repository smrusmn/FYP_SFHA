```python
import spacy
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
nlp = spacy.load("en_core_web_sm")
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Sub2Sys_string.csv')
df.head()
```

```python
df.isnull().any()
```

```python
df.columns = df.columns.str.replace(' ', '_')
df = df.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
df.columns = df.columns.str.replace('[^\w\s]', '')
df = df.applymap(lambda x: ''.join(e for e in str(x) if e.isalnum() or e.isspace()) if isinstance(x, str) else x)
df.head()
```

```python
X = df.drop('DAL', axis = 'columns')
X.head()
y = df.DAL
y.head()
```

```python
df.columns
```

```python
docs = [nlp(text) for text in (df['Sub_System'])]
# print(docs)
x1_vecs = []
trf = spacy.load('en_core_web_sm')
for doc in trf.pipe(df['Sub_System']):
  a = doc.vector
  x1_vecs.append(a)
print(docs)
# print(x1_vecs)
```

```python
docs = [nlp(text) for text in (df['Failure_Condition_1s'])]
# print(docs)
x2_vecs = []
trf = spacy.load('en_core_web_sm')
for doc in trf.pipe(df['Failure_Condition_1s']):
  a = doc.vector
  x2_vecs.append(a)
print(docs)
# print(x1_vecs)
```

```python
docs = [nlp(text) for text in (df['Failure_Condition_2s'])]
# print(docs)
x3_vecs = []
trf = spacy.load('en_core_web_sm')
for doc in trf.pipe(df['Failure_Condition_2s']):
  a = doc.vector
  x3_vecs.append(a)
print(docs)
# print(x1_vecs)
```

```python
docs = [nlp(text) for text in (df['Parent_Condition_1s'])]
# print(docs)
x4_vecs = []
trf = spacy.load('en_core_web_sm')
for doc in trf.pipe(df['Parent_Condition_1s']):
  a = doc.vector
  x4_vecs.append(a)
print(docs)
# print(x1_vecs)
```

```python
docs = [nlp(text) for text in (df['Parent_Condition_2s'])]
# print(docs)
x5_vecs = []
trf = spacy.load('en_core_web_sm')
for doc in trf.pipe(df['Parent_Condition_2s']):
  a = doc.vector
  x5_vecs.append(a)
print(docs)
# print(x1_vecs)
```

```python
docs = [nlp(text) for text in (df['DALs'])]
# print(docs)
x6_vecs = []
trf = spacy.load('en_core_web_sm')
for doc in trf.pipe(df['DALs']):
  a = doc.vector
  x6_vecs.append(a)
print(docs)
# print(x1_vecs)
```

```python
df.head()
```

```python
x1_series = pd.Series(x1_vecs, name='Sub_System')
x2_series = pd.Series(x2_vecs, name='Failure_Condition_1s')
x3_series = pd.Series(x3_vecs, name='Failure_Condition_2s')
x4_series = pd.Series(x4_vecs, name='Parent_Condition_1s')
x5_series = pd.Series(x5_vecs, name='Parent_Condition_2s')
x6_series = pd.Series(x6_vecs, name='DALs')
X_vecs = pd.concat([x1_series, x2_series, x3_series, x4_series, x5_series, x6_series], axis = 1)

X_vecs.head()
```

```python
X_vecs.isnull().sum()
```

```python
X_vecs.shape
```

```python
def vector_to_constant(vec):
    vec_array = np.array(vec)
    magnitude = np.linalg.norm(vec_array)
    return magnitude
```

```python
X_vecs1 = X_vecs.copy()
for column in X_vecs1.columns:
    # Apply the vector_to_constant function to each entry in the column
    X_vecs1[column] = X_vecs[column].apply(vector_to_constant)
X_vecs1.head()
```

```python
# X_vecs1.to_csv('X_mag.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_vecs1,y,test_size = 0.2, random_state = 42,shuffle = True)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
```

```python
clf2 = LogisticRegression()
clf2.fit(X_train,y_train)
```

```python
clf3 = DecisionTreeClassifier()
clf3.fit(X_train,y_train)
```

```python
y_result = clf.predict(X_test)
print(y_result)
```

```python
y_result = clf2.predict(X_test)
print(y_result)
```

```python
y_result = clf3.predict(X_test)
print(y_result)
print(y_test)
```

```python
accuracysvm = clf.score(X_test, y_test)
print(f'Test Accuracy: {accuracysvm}')
```

```python
accuracyLR = clf2.score(X_test, y_test)
print(f'Test Accuracy: {accuracyLR}')
```

```python
accuracyDT = clf3.score(X_test, y_test)
print(f'Test Accuracy: {accuracyDT}')
```
