```python
import pandas as pd
```

```python
df = pd.read_csv('Sub2Sys_strings.csv')
df.head()
```

```python
df['DAL'] = df.DAL.apply(lambda x: 'Catastrophic' if x==4 else x)
df.head()
```

```python
df['DAL'] = df.DAL.apply(lambda x: 'Hazardous' if x==3 else x)
df.head()
```

```python
df['DAL'] = df.DAL.apply(lambda x: 'Major' if x==2 else x)
df.head()
```

```python
df['DAL'] = df.DAL.apply(lambda x: 'Minor' if x==1 else x)
df.head()
```

```python
df['DAL'] = df.DAL.apply(lambda x: 'None' if x==0 else x)
df.head()
```

```python
df.to_csv('Sub2Sys_string.csv',index=False)
```
